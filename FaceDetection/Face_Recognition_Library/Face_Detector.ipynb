```
import numpy as np
import cv2
import face_recognition as fr
from google.colab.patches import cv2_imshow
```

```
image_obama = fr.load_image_file('images/obama_train2.jpeg')
image_obama_copy = cv2.cvtColor(image_obama, cv2.COLOR_BGR2RGB)
```

```
cv2_imshow(image_obama_copy)
```

```
image_obama_test = fr.load_image_file('images/obama_test.jpeg')
image_obama_test_copy = cv2.cvtColor(image_obama_test, cv2.COLOR_BGR2RGB)
```

```
cv2_imshow(image_obama_test_copy)
```

```
face_location = face_recognition.face_locations(image_obama_copy)[0]
encode_obama = face_recognition.face_encodings(image_obama_copy)[0]
cv2.rectangle(image_obama_copy, (face_location[3], face_location[0]), (face_location[1], face_location[2]), (255, 0, 255), 2)
cv2_imshow(image_obama_copy)
```

```
face_location_test = face_recognition.face_locations(image_obama_test_copy)[0]
encode_obama_test = face_recognition.face_encodings(image_obama_test_copy)[0]
cv2.rectangle(image_obama_test_copy, (face_location_test[3], face_location_test[0]), (face_location_test[1], face_location_test[2]), (255, 0, 255), 2)
cv2_imshow(image_obama_test_copy)
```

```
results = face_recognition.compare_faces([encode_obama], encode_obama_test)
face_distance = face_recognition.face_distance([encode_obama], encode_obama_test)
# print(results)
# print(face_distance)
cv2.putText(image_obama_test_copy, f'{results[0]} {round(face_distance[0], 2)}', (50, 50), cv2.FONT_HERSHEY_COMPLEX, .7, (0, 0, 255), 2)
cv2_imshow(image_obama_test_copy)
```
